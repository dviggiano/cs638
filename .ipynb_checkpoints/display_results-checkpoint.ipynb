```python
import pandas as pd
import os

# List all CSV files in the directory
csv_files = [f'results/{i}.csv' for i in range(1, 33)]

# Process each file
for file in csv_files:
    # Read the CSV file
    df = pd.read_csv(file)
    
    # Check if the 'prompt' column exists, then remove it
    if 'Prompt' in df.columns:
        df = df.drop('prompt', axis=1)
    
    # Print the DataFrame
    print(f"{file}:")
    print(df)
    print("\n")
```
